```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Lorentz system equations
def lorentz_system(x, y, z, sigma, rho, beta):
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return dx, dy, dz

# Define the parameters for the simulation
sigma = 10
rho = 28
beta = 8/3

# Define the initial conditions
x0 = 1
y0 = 1
z0 = 1

# Define the time step and number of steps
dt = 0.01
num_steps = 1000

# Initialize the arrays to store the results
x = np.zeros(num_steps)
y = np.zeros(num_steps)
z = np.zeros(num_steps)

# Set the initial conditions
x[0] = x0
y[0] = y0
z[0] = z0

# Run the simulation
for i in range(1, num_steps):
    dx, dy, dz = lorentz_system(x[i-1], y[i-1], z[i-1], sigma, rho, beta)
    x[i] = x[i-1] + dx * dt
    y[i] = y[i-1] + dy * dt
    z[i] = z[i-1] + dz * dt

# Plot the results
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
plt.show()
```
